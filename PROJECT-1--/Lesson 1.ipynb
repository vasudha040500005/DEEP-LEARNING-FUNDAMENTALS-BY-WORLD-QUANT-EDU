```python
## Linear Functions in 1D and 2D, and Their Use in Machine Learning and Deep Learning

### Linear Function in 1D
A linear function in one dimension (1D) is a straight-line relationship between two variables, typically expressed as:
 y = mx + c
where:
-  m  is the slope (rate of change),
-  c  is the y-intercept (value of y when x=0),
-  x  is the input variable,
-  y  is the output variable.

For example, with parameters  c=10  and  m=-2 , the function is  y = -2x + 10.

### Linear Function in 2D
In two dimensions (2D), a linear function represents a plane. For two input variables  x_1  and  x_2 , it can be written as:
 y = w_1 x_1 + w_2 x_2 + b 
where:
-  w_1  and  w_2  are weights (coefficients),
-  b  is the bias term,
-  y  is the output.

This extends to higher dimensions, like 11D, where  y = \sum_{i=1}^{11} w_i x_i + b , representing a hyperplane.

### Use in Machine Learning and Deep Learning
- **Machine Learning**: Linear functions form the basis of linear regression, where we fit a line (or plane) to data to predict outcomes. They are simple, interpretable, and used for tasks like predicting house prices based on features.
- **Deep Learning**: In neural networks, linear functions are implemented as fully connected layers (dense layers). Each neuron computes a weighted sum of inputs plus bias, followed by an activation function (e.g., ReLU). This allows stacking layers to learn complex patterns, as in feedforward networks for classification or regression. For instance, in a simple perceptron, it's a linear classifier.
```

```python

#To calculate the slope of a line we need two parameters m and c


import matplotlib.pyplot as plt
import numpy as np

# Parameters
c = 10
m = 2

# Create x values
x = np.linspace(-5, 10, 100)

# Calculate y values using the linear function y = mx + c
def linear_func(x,m,c):
    return  m * x + c

y=linear_func(x,m,c)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, linewidth=2, label=f'y = {m}x + {c}')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Linear Function in 1D', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```
